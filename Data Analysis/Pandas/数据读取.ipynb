```python
# This function returns the hundredth item from a series
def hundreth_row(column):
    # Extract the hundredth item
    hundredth_item = column.loc[99]
    return hundredth_item

# Return the hundredth item from each column
hundredth_row = data.apply(hundreth_row)
print(hundredth_row)
```

```python
def not_null_count(column):
    column_null = pandas.isnull(column)
    null = column[column_null]
    return len(null)

column_null_count = data.apply(not_null_count)
print(column_null_count)
```

```python
def is_Pn(row):
    if row['C3 Adjustment'] > 0:
        return 'Positive Number'
    else:
        return 'Negitive Number'
sub = data.apply(is_Pn, axis=1)
print(sub)
```
