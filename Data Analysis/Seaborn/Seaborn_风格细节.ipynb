```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

# seaborn: statistical data visualization

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sinplot()
```

```python
sns.set()
sinplot()
```

## 5种主题风格
- darkgrid
- whitegrid
- dark
- white
- ticks

```python
sns.set_style('whitegrid')
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data)
```

```python
sns.set_style('dark')
sinplot()
```

```python
sns.set_style('ticks')
sinplot()
```

```python
sns.set_style('white')
sinplot()
sns.despine(left=True)
```

```python
sns.violinplot(data)
sns.despine(offset=10)
```

```python
# 用with打开一种风格
with sns.axes_style('darkgrid'):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot()
```

```python
sns.set_context('paper')
plt.figure(figsize=(10, 6))
sinplot()
```

```python
sns.set_context('talk')
plt.figure(figsize=(10, 6))
sinplot()
```

```python
sns.set_context('notebook', font_scale=1.7, rc={'lines.linewidth':2.4})
plt.figure(figsize=(10, 6))
sinplot()
```
