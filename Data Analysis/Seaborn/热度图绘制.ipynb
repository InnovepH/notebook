```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(0)
import seaborn as sns
sns.set()
```

```python
uniform_data = np.random.rand(3, 3)
print(uniform_data)
heatmap = sns.heatmap(uniform_data)
```

```python
uniform_data = np.random.rand(4, 4)
print(uniform_data)
heatmap = sns.heatmap(uniform_data)
```

```python
ax = sns.heatmap(uniform_data, vmin=0.2, vmax=0.5)
```

```python
normal_data = np.random.randn(4, 4)
print(normal_data)
ax = sns.heatmap(normal_data, center=0)
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights_pivot = flights.pivot('month', 'year', 'passengers')
print(flights_pivot)
ax = sns.heatmap(flights_pivot)
```

```python
ax = sns.heatmap(flights_pivot, annot=True, fmt='d')
```

```python
ax = sns.heatmap(flights_pivot, linewidths=.2)
```

```python
ax = sns.heatmap(flights_pivot, cmap='YlGnBu')
```

```python
ax = sns.heatmap(flights_pivot, cbar=False)
```
